PYTHON Koşul Yapıları

if Koşul yapısı -> belirtilen şart doğruysa (True) bir bloğun çalıştırılmasını sağlamaktadır. if-else Koşul yapısı -> if bloğu sağlanmazsa, else bloğu çalıştırılır. if-elif-else Koşul yapısı -> Birden fazla koşul kontrol etmek için elif (else if) kullanılır. İlk doğru koşul bulunana kadar kontrol devam eder. Hiçbiri doğru değilse, else bloğu çalıştırılır. İç İçe Koşullar Mantıksal Operatörler ile Koşul Yapıları

```python
try:
  a=int(input("lütfen bir sayi giriniz"))

  if a>3:
    print("a 3 den büyüktür")
  elif  a==3 :
    print("a 3 e eşittir")
  else:
    print("a3 den küçüktür")
except ValueError:
  print("doğru sayı giriniz")
```

```python
# iif koşulu

x=5

if x>3:
  print(" x 5 den büyüktür")
```

```python
# iif else koşulu

x=5

if x>3:
  print(" x 3 den büyüktür")
else :
  print("x 3 den küçüktür")
```

```python
x=10

if x>10:
  print("x10 den büyük")
elif x>5:
  print("x 5den büyüktür ama 10 dan küçük veya eşittir")
else :
  print("x 5 den küçük veya eşittir")
```

```python
#iç içe
x=12
y=10

if x>10:
  if y>8:
    print("x 10 dan büyük ve y 8 den büyüktür")
  else:
    print("x 10 dan büyük ama y 8den küçük veya eşit")
else:
  print("x 10 dan küçüktür")
```

```python
# Mantıksal Operatörler ile Koşullar
x=7
y=3

if x>5 and y>2:
  print("x,5 den büyük ve y 2den büyüktür")

if x>5 or y>5:
  print("x veya y 5 den büyüktür")

if not (x>10):  #10dan büüyk değilse
  print("x, 10 dan büyük değildir")
```

```python
# Örnek 1: Not Sistemi   BAKILACAK
try:
  ogr_not=int(input("Notunuzu giriniz: "))
  if ogr_not>=85 and ogr_not<=100:
      print("A")
  elif ogr_not >=80 and ogr_not<85:
      print("B")
  elif ogr_not >=70 and ogr_not<80:
      print("C")
  elif ogr_not >=60 and ogr_not<70:
      print("D")
  elif ogr_not<0 or ogr_not >100:
      print("geçerli not giriniz")
  else:
      print("F")

except ValueError:
  print("Geçerli not giriniz")
```

```python
# Örnek 2: Yaş Kontrolü

yas = 10

if yas < 18:
    print("Reşit değilsiniz.")
elif yas < 21:
    print("Reşitsiniz ama içecek satın alamazsınız.")
else:
    print("İçecek satın alabilirsiniz.")
```
