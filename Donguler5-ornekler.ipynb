```python
# 1 ile 100 arasındaki sayılardan 7'ye bölünenleri bulan Python kodu
sayi_dizisi=[]
sayi=1
while sayi<=100:
  if sayi%7==0:
    sayi_dizisi.append(sayi)
  sayi+=1
print("1 ile 100 arasında 7 ye bölünen sayılar",sayi_dizisi)
print(f"1ile 100 arasında 7 ye bölünen kaç sayı var--> {len(sayi_dizisi)}")
```

```python
# Kullanıcıdan alınan sayıya kadar 1'den başlayarak faktöreyel hesabının yapılmasıdır.
while True:
    try:
        sayi = int(input("Bir sayı girin: "))
        if sayi < 0:
            print("Negatif sayıların faktöriyeli hesaplanamaz.")
        elif sayi == 0 or sayi == 1:
            print(1)
        else:
            sonuc = 1
            while sayi > 1:
                sonuc *= sayi
                sayi -= 1
            print(sonuc)
        break
    except ValueError:
        print("Lütfen geçerli bir tam sayı girin.")
```

```python
#Kullanıcıdan başlangıç ve bitiş değerlerini alarak, bu değerler arasındaki çift sayıları bulan ve bu çift sayıların toplamını hesaplayan bir Python programı yazınız.

# Kullanıcıdan başlangıç değeri al
while True:
  try:
    baslangic_deger=int(input("1. sayıyı giriniz"))
    if baslangic_deger<0:
      baslangic_deger= abs(baslangic_deger)
      print("değer negatif olamaz + ya cevrildi")
    break
  except ValueError:
    print("Geçerli bir değer girnizi")

# Kullanıcıdan bitiş değeri al

while True:
  try:
    bitis_deger=int(input("2. sayıyı giriniz"))
    if bitis_deger<0:
      bitis_deger= abs(bitis_deger)
      print("değer negatif olamaz + ya cevrildi")
    break
  except ValueError:
    print("Geçerli bir değer girnizi")

# Başlangıç ve bitiş değerleri arasında çift sayıları bul ve topla
cift_sayi=[]
cift_topla=0
sayi=baslangic_deger
while sayi<=bitis_deger:
  if sayi % 2==0:
    cift_sayi.append(sayi)
  sayi+=1
print(f"{baslangic_deger} ve {bitis_deger} arasındaki çift sayılar: ",cift_sayi)

# Elde edilen çift sayıları ve toplamlarını ekrana yazdır
print(f"{baslangic_deger} ve {bitis_deger} arasındaki çift sayılar: ")
for num in cift_sayi:
  cift_topla+=num
  print(num,end=",")
print("\nÇİFT SAYILARIN TOPLAMI-->",cift_topla)


```

```python
# Önceden tanımlalan sesli harf listesinde girilen metnin karşılaştırılması yapılarak sesli ve sessiz harf kontrolünün yapımı

sesli_harf="aeıioöuüAEIİOÖUÜ"

kelime=input("bir kelime giriniz")

# Sesli ve sessiz harf sayısını hesapla
sesli_harf_sayisi=0
sessiz_harf_sayisi=0

#isalpha() ile bir stringde yalnızca harf olup olmadığını kontrol etmek için kullanılır.

#Aşağıdaki kod blogunda geçen "isalpha" harf dışı bir karakter ile karşılaşınca "false" döndürmektedir.

for harf in kelime:
    if harf.isalpha():
      if harf in sesli_harf: #karşılaştırma
        sesli_harf_sayisi+=1
      else:
        sessiz_harf_sayisi+=1
print(kelime," kelimseinde ",sesli_harf_sayisi," adet sesli harf ",sessiz_harf_sayisi," adet sessiz harf mevcuttur")
```
